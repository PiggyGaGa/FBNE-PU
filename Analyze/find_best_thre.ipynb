```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
# before this process, we caculate the probability of the nnPU outputs in our dataset.

TaxZ_Dis = [52795, 3, 3, 0, 0, 0, 209, 40, 2, 4390]
TaxH_Dis = [72073, 1, 2, 0, 0, 0, 0, 0,360, 1490]
TaxS_Dis = [25227, 5533, 3735, 1477, 1268, 1116, 1198, 1150, 1423, 10108]

TaxZ_Dis_Norm = np.array(TaxZ_Dis) / sum(TaxZ_Dis)

TaxH_Dis_Norm = np.array(TaxH_Dis) / sum(TaxH_Dis)

TaxS_Dis_Norm = np.array(TaxS_Dis) / sum(TaxS_Dis)
```

```python
def plot_bar(value1, value2, value3, legend=['a', 'b', 'c']):
    width = 0.5
    label = ['[0, 0.1)', '[0.1, 0.2)', '[0.2, 0.3)', '[0.3, 0.4)', '[0.4, 0.5)', '[0.5, 0.6)', '[0.6, 0.7)', '[0.7, 0.8)', '[0.8, 0.9)', '[0.9, 1]']
    x = (np.array(range(len(value1)))+1) * 3
    fig, ax = plt.subplots()
    ax.set_xticks(x+ width/2)
    ax.set_xticklabels(label, fontsize=6)
    ax.bar(x, value1, width,alpha = 0.9, color='#F1B94A')
    ax.bar(x+width, value2, width,alpha = 0.9, color='#0D75DC')
    ax.bar(x+2*width, value3, width,alpha = 0.9, color='#7CBF7B')
    # ax.plot(x, value)
    plt.legend(legend)
    plt.xlabel('Interval')
    plt.ylabel('Probability Distribution')
    plt.savefig('ProbabilityDistribution.pdf')
```

```python
plot_bar(TaxS_Dis_Norm, TaxH_Dis_Norm, TaxZ_Dis_Norm, legend=['TaxS', 'TaxH', 'TaxZ'])
```
